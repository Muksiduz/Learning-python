```python
import pdfplumber
import re
import pandas as pd
```

```python
def extract_the_pdf_data(pdf_path):
    all_data = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            all_data.append(text)
    # print(all_data)
    return all_data
def extract_transaction_data(raw_text):
    transactions = []
    one_transactions = {}
    # Define regex patterns to capture date, paid to, amount, time, and paid by
    date_pattern = r"([A-Za-z]{3} \d{1,2}, \d{4}) Paid to"
    paid_to_pattern = r"([A-Za-z]{3} \d{2}, \d{4}) (Transfer to|Paid to) (.+)"
    amount_pattern = r"₹([\d,]+)"
    time_pattern = r"(\d{2}:\d{2} [ap]m)"
    paid_by_pattern = r"Paid by (\w+)"

    for lines in raw_text:
        print(lines)

        # Extract date using regex pattern
        date_match = re.search(date_pattern, lines)
        if date_match:
            date = date_match.group(1)
            one_transactions["Date"] = date

        # Extract paid to using regex pattern
        paid_to_match = re.search(paid_to_pattern, lines)
        if paid_to_match:
            paid_to = paid_to_match.group(1)
            one_transactions["Paid to"] = paid_to

        # Extract amount using regex pattern
        amount_match = re.search(amount_pattern, lines)
        if amount_match:
            amount = amount_match.group(1)
            one_transactions["Amount"] = amount

        # Extract time using regex pattern
        time_match = re.search(time_pattern, lines)
        if time_match:
            time = time_match.group(1)
            one_transactions["Time"] = time

        # Extract paid by using regex pattern
        paid_by_match = re.search(paid_by_pattern, lines)
        if paid_by_match:
            paid_by = paid_by_match.group(1)
            one_transactions["Paid by"] = paid_by
        transactions.append(one_transactions)
        one_transactions = {}
    return transactions


transactions = extract_transaction_data(extract_the_pdf_data("./transaction.pdf"))

df = pd.DataFrame(transactions)
```
